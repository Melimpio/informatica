```python
# Bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
import numpy as np

# Carregando os dados
df = pd.read_excel("dados_imoveis_regressao.ipynb.xlsx")

# Selecionando apenas as colunas relevantes e removendo valores ausentes
df = df[['Área', 'Preço']].dropna()

# Definindo as variáveis
X = df[['Área']]
y = df['Preço']

# Criando e ajustando o modelo de regressão linear
modelo = LinearRegression()
modelo.fit(X, y)

# Fazendo previsões
y_pred = modelo.predict(X)

# Gráfico com seaborn (reta de regressão + dispersão dos pontos)
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Área', y='Preço', data=df, color='blue', label='Dados reais')
plt.plot(df['Área'], y_pred, color='red', linewidth=2, label='Regressão Linear')
plt.title('Regressão Linear: Preço vs Área')
plt.xlabel('Área (m²)')
plt.ylabel('Preço (mil R$)')
plt.legend()
plt.grid(True)
plt.show()

# Exibindo coeficientes da regressão
print(f"Coeficiente angular (inclinação): {modelo.coef_[0]:.2f}")
print(f"Coeficiente linear (intercepto): {modelo.intercept_:.2f}")
print(f"R² (coeficiente de determinação): {modelo.score(X, y):.4f}")
```
